```python
#importing neccessary libraries
import pandas as pd
import numpy as np
```

```python
#loading the energy dataset 
df = pd.read_csv("energydata_complete.csv")
```

```python
#displaying first 5 rows of dataset
df.head()
```

```python
"""checking the general information of the data like 
 dataset lenght,Number of columns,datatypes of each columns 
"""
df.info()
```

```python
#Check Missing values
'''
This function calculates the feature with the missing values,
Total number of nans
percentage of the missing values in dataset

'''
def Missing_Values(df):
    Missing_Data=df.isnull().sum().reset_index().rename(columns={'index':'Fuel_label',0:'Missing_Values'})
    Feature_Value=Missing_Data[Missing_Data['Missing_Values']==Missing_Data['Missing_Values'].max()]['Fuel_label'].tolist()[0]
    Total_Value=Missing_Data[Missing_Data['Missing_Values']==Missing_Data['Missing_Values'].max()]['Missing_Values'].tolist()[0]
    Percent_Value=100*(Missing_Data[Missing_Data['Missing_Values']==Missing_Data['Missing_Values'].max()]['Missing_Values'].tolist()[0]/len(df))
    print('Feature:',Feature_Value ,',', f'Total:{Total_Value:.3f}',',',f'Percent:{Percent_Value:.3f}')

Missing_Values(df) 
```

```python
#dropping some columns Lights and date
df = df.drop(columns=['lights', 'date'])
```

```python
#Checking the above columns if it was dropped
df.info()
```

```python
#import MinMaxScaler for scaling the dataset
'''
This ensures the dataset is well scaled.
'''
from sklearn.preprocessing import MinMaxScaler
sc= MinMaxScaler()
df_normalized= pd.DataFrame(sc.fit_transform(df), columns = df.columns)
```

```python
#Getting and Droping the target varriable
'''
    Here we separate the target(Appliances) from the rest varriables
'''
X =df_normalized.drop(columns=['Appliances'])
y =df_normalized['Appliances']
```

```python
#Now, we split our dataset into the training and testing dataset.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=42)
```

```python
#Creating the Linear multiple Linear regression Model
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
#fitting the model to the training Dataset
linear_model.fit(x_train, y_train)
#obtaining predictions
predicted_values = linear_model.predict(x_test)
```

```python
#Question 13
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
def get_weights_df(model, feat, col_name):
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
```

```python
#Question 11
import pandas as pd
import numpy as np
```

```python
#loading the energy dataset 
df = pd.read_csv("energydata_complete.csv")
```

```python
df = df.drop(columns=['date', 'lights'])
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
X = normalised_df[['T2']]
y = normalised_df[['T6']]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3, random_state=42)
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)

predicted_values = linear_model.predict(x_test)
```

```python
Question 11
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```
